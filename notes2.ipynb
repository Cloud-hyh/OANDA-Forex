#Timeseries And Quant Research on Historical Data

```python
%matplotlib inline
from timeseries import *
```

```python
q1 = EventQueue()
q2 = EventQueue()
q = {'mkt': q1, 'bar': q2}

api = PyApi(ResearchConfig(), q) # Use ResearchConfig() object to construct, rather than Config()
```

## Get Data method

* timeseries.get_data(api, count, granularity, instrument)

```python
data = get_data(api, count=500, granularity='M30', instrument='EUR_USD')
data.head()
```

```python
data.tail()
```

## Plot method

* timeseries.tsplot(data, ma1, ma2, save, name, hf)

```python
tsplot(data,save=0)
```

```python
tsplot(get_data(api, count=1500, granularity='D', instrument='USD_JPY'),
       ma1=50, ma2=150,save=0)
```

```python
tsplot(get_data(api, count=5000, granularity='M5', instrument='USD_JPY'), 
       ma1=100, ma2=200,save=0)
```

## High Frequency Data Plot Setting

* timeseries.tsplot(data, ma1, ma2, save, name, hf=1)

```python
tsplot(get_data(api, count=200, granularity='M1', instrument='EUR_USD'),save=0,hf=1)
```
